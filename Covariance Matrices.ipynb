```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_bivariate_data(size, prob, cov=[], mew=[]):
    m1 = mew[0]
    cov1 = cov[0]
    C1 = np.random.multivariate_normal(m1, cov1, size)
    data = C1
    labels = np.hstack(np.ones(len(C1)))
    return data, labels
```

```python


plt.figure(figsize=(6,4))


plt.subplot(221)
data, labels = generate_bivariate_data(1000, 0.5, [[[4,0],[0,4]]], [[0,0]])
plt.plot(data[labels == 1,0],data[labels == 1,1],'.',color='k', alpha=0.5)
plt.grid('on')
plt.axis('Equal')
plt.title('[[4,0], [0,4]]')

plt.subplot(222)
data, labels = generate_bivariate_data(1000, 0.5, [[[8,0],[0,1]]], [[0,0]])
plt.plot(data[labels == 1,0],data[labels == 1,1],'.',color='k', alpha=0.5)
plt.grid('on')
plt.axis('Equal')
plt.title('[[8,0], [0,1]]')

plt.subplot(223)
data, labels = generate_bivariate_data(1000, 0.5, [[[4,-4],[-4,6]]], [[0,0]])
plt.plot(data[labels == 1,0],data[labels == 1,1],'.',color='k', alpha=0.5)
plt.grid('on')
plt.axis('Equal')
plt.title('[[4,-4], [-4,6]]')

plt.subplot(224)
data, labels = generate_bivariate_data(1000, 0.5, [[[4,4],[4,6]]], [[0,0]])
plt.plot(data[labels == 1,0],data[labels == 1,1],'.',color='k', alpha=0.5)
plt.grid('on')
plt.axis('Equal')
plt.title('[[4,4], [4,6]]')

plt.show()

```
